```python
# Importing Essential Packages and Loading the data set into a data frame.

import numpy as np
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from statistics import mean 
import warnings
warnings.filterwarnings("ignore")

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score


df = pd.read_excel('DataSetPredictions.xlsx', sheet_name='Sheet2')
df.head()
```

```python
# Create a list to capture Polarity of the the comments
Polarity = []
# Import TextBolb Module for Sentiment Anslysis
from textblob import TextBlob
# Loop through the comments to derive Polarity and append to list
for i in range(len(df.Comment)):
     Polarity.append(TextBlob(df.Comment[i]).sentiment.polarity)
# Add the new column of polarity to the Data Frame
df['Polarity'] = Polarity
df.head()
```

```python
df['Normal Polarity'] = (df.Polarity+1)/2
df.head()
```

```python
# Replacing All NNeutral Reviews as Bad Reviews.

df = df.replace('Neutral','Bad Review')
```

```python
# Encoding the Lables
# Import label encoder 
from sklearn import preprocessing 
  
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
df['Classification']= label_encoder.fit_transform(df['Classification']) 
df.head()
```

```python
#Get a count of classification
df.Classification.value_counts()
```

```python
# library & dataset
import seaborn as sns
# Use the 'hue' argument to provide a factor variable
sns.lmplot( x="Average_Rating", y="Normal Polarity", data=df, fit_reg=False, hue='Classification', legend=False)
```

```python
# Extracting X and Y which are Input and Target variables for Random Forest
# Target Variables
Y = df['Classification']
Y.head()
```

```python
X = df[['Average_Rating','Normal Polarity','Comment']]
X.head()
```

```python

# Splitting Data into Train and Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.1,random_state = 45)

# Logistic Regression Classification Model


lr = LogisticRegression()
lr.fit(X_train[['Average_Rating','Normal Polarity']],y_train)
lr_prediction = lr.predict(X_test[['Average_Rating','Normal Polarity']])
Logistic_Regression_Accuracy = accuracy_score(y_test, lr_prediction)
print('Logistic Regression Accuracy :',Logistic_Regression_Accuracy)
lr_scores = cross_val_score(lr, X_train[['Average_Rating','Normal Polarity']],y_train, cv=10, scoring='accuracy')
print('Cross Validation Logistic Regression Accuracy :', mean(lr_scores))

# Naive Bayes Classification Model


nb = GaussianNB()
nb.fit(X_train[['Average_Rating','Normal Polarity']],y_train)
nb_prediction = nb.predict(X_test[['Average_Rating','Normal Polarity']])
Naieve_Bayes_Accuracy = accuracy_score(y_test, nb_prediction)
print('Naive Bayes Accuracy :',Naieve_Bayes_Accuracy)
nb_scores = cross_val_score(nb, X_train[['Average_Rating','Normal Polarity']],y_train, cv=10, scoring='accuracy')
print('Cross Validation Naive Bayes Accuracy :',mean(nb_scores))

# Stochastic Gradient Descent


sgd = SGDClassifier(loss = 'modified_huber', shuffle = True, random_state = 5)
sgd.fit(X_train[['Average_Rating','Normal Polarity']],y_train)
sgd_prediction = sgd.predict(X_test[['Average_Rating','Normal Polarity']])
Stochastic_Gradient_Descent_Accuracy = accuracy_score(y_test, sgd_prediction)
print('Stochastic Gradient Descent Accuracy :',Stochastic_Gradient_Descent_Accuracy)
sgd_scores = cross_val_score(sgd, X_train[['Average_Rating','Normal Polarity']],y_train, cv=10, scoring='accuracy')
print('Cross Validation Stochastic Gradient Descent Accuracy Accuracy :',mean(sgd_scores))

# K Nearest Neighbours


# for i in range(1,20):
#     knn = KNeighborsClassifier(n_neighbors = i)
#     knn.fit(X_train[['Average_Rating','Normal Polarity']],y_train)
#     knn_prediction = knn.predict(X_test[['Average_Rating','Normal Polarity']])
#     K_Nearest_Neighbors_Accuracy = accuracy_score(y_test, knn_prediction)
#     print('K Nearest Neighbours Accuracy for k :',i,'is : ',K_Nearest_Neighbors_Accuracy) 
#     knn_scores = cross_val_score(knn, X_train[['Average_Rating','Normal Polarity']],y_train, cv=10, scoring='accuracy')
#     print('Cross Validation K Nearest Neighbours Accuracy for k :',i,'is : ',mean(knn_scores))

# Let us run the Knn for K = 1

knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train[['Average_Rating','Normal Polarity']],y_train)
knn_prediction = knn.predict(X_test[['Average_Rating','Normal Polarity']])
K_Nearest_Neighbors_Accuracy = accuracy_score(y_test, knn_prediction)
print('K Nearest Neighbours Accuracy for k :',i,'is : ',K_Nearest_Neighbors_Accuracy) 
knn_scores = cross_val_score(knn, X_train[['Average_Rating','Normal Polarity']],y_train, cv=10, scoring='accuracy')
print('Cross Validation K Nearest Neighbours Accuracy for k :',i,'is : ',mean(knn_scores))

# Random Forest Classifier Implementation

from sklearn.metrics import accuracy_score
from statistics import mean 
rf = RandomForestClassifier(n_estimators=100, max_depth=5, random_state = 5)
rf.fit(X_train[['Average_Rating','Normal Polarity']],y_train)
rf_prediction = rf.predict(X_test[['Average_Rating','Normal Polarity']])
Random_Forest_Accuracy = accuracy_score(y_test, rf_prediction)
print('Random Forest Accuracy for :',Random_Forest_Accuracy)  
rf_scores = cross_val_score(rf, X_train[['Average_Rating','Normal Polarity']],y_train, cv=10, scoring='accuracy')
print('Cross Validation Random Forest Accuracy ',mean(rf_scores))

# Support Vector Machine Implementation -- Linear kernel

svm_lin = SVC(kernel='linear', C = 0.025, random_state = 6)
svm_lin.fit(X_train[['Average_Rating','Normal Polarity']],y_train)
svm_lin_prediction = svm_lin.predict(X_test[['Average_Rating','Normal Polarity']])
Support_Vector_Accuracy_lin = accuracy_score(y_test, svm_lin_prediction)
print('Support Vector Accuracy for Linear Kernel :',Support_Vector_Accuracy_lin)  
svm_lin_scores = cross_val_score(svm_lin, X_train[['Average_Rating','Normal Polarity']],y_train, cv=10, scoring='accuracy')
print('Cross Validation Support Vector Accuracy for Linear Kernel ',mean(svm_lin_scores))

# Support Vector Machine Implementation -- Poly kernel

svm_poly = SVC(kernel='poly', C = 0.025, random_state = 6)
svm_poly.fit(X_train[['Average_Rating','Normal Polarity']],y_train)
svm_poly_prediction = svm_poly.predict(X_test[['Average_Rating','Normal Polarity']])
Support_Vector_Accuracy_poly = accuracy_score(y_test, svm_poly_prediction)
print('Support Vector Accuracy for Ploynomial Kernel :',Support_Vector_Accuracy_poly)  
svm_poly_scores = cross_val_score(svm_poly, X_train[['Average_Rating','Normal Polarity']],y_train, cv=10, scoring='accuracy')
print('Cross Validation Support Vector Accuracy for Ploynomial Kernel ',mean(svm_poly_scores))

# Support Vector Machine Implementation -- rbf kernel

svm_rbf = SVC(kernel='rbf', C = 0.025, random_state = 6)
svm_rbf.fit(X_train[['Average_Rating','Normal Polarity']],y_train)
svm_rbf_prediction = svm_rbf.predict(X_test[['Average_Rating','Normal Polarity']])
Support_Vector_Accuracy_rbf = accuracy_score(y_test, svm_rbf_prediction)
print('Support Vector Accuracy for rbf Kernel :',Support_Vector_Accuracy_rbf)  
svm_rbf_scores = cross_val_score(svm_rbf, X_train[['Average_Rating','Normal Polarity']],y_train, cv=10, scoring='accuracy')
print('Cross Validation Support Vector Accuracy for rbf Kernel ',mean(svm_rbf_scores))

# Support Vector Machine Implementation -- sigmoid kernel

svm_sig = SVC(kernel='sigmoid', C = 0.025, random_state = 6)
svm_sig.fit(X_train[['Average_Rating','Normal Polarity']],y_train)
svm_sig_prediction = svm_sig.predict(X_test[['Average_Rating','Normal Polarity']])
Support_Vector_Accuracy_sig = accuracy_score(y_test, svm_sig_prediction)
print('Support Vector Accuracy for Sigmoid Kernel :',Support_Vector_Accuracy_sig)  
svm_sig_scores = cross_val_score(svm_sig, X_train[['Average_Rating','Normal Polarity']],y_train, cv=10, scoring='accuracy')
print('Cross Validation Support Vector Accuracy for Sigmoid Kernel ',mean(svm_sig_scores))
```
